### Ratio financiers###
**Interest ratios :**

EBIT Interest Coverage = Earnings Before Interest & Taxes / Interest Expense

EBITDA Interest Coverage = Earnings Before Interest, Taxes, Depreciation & Amortization / Interest Expense

**Leverage ratios :** 

Debt to Equity = (Current Debt + Long Term Debt) / Common Equity(ou stockholder equity)

Debt to Tangible Equity = (Current Debt + Long Term Debt) / (Common Equity - Intangible Assets)

Cash Flow to Total Debt = Cash Flow from Operations / Total Debt

Liabilities to Assets = Total Liabilities / Total Assets

**Liquidity ratios :**

Cash Ratio = (Cash & Short Term Investments) / Current Liabilities

Quick Ratio = (Cash & Short Term Investments + Receivables) / Current Liabilities

Current Ratio = Current Assets / Current Liabilitie


```python
dict_df_ratios = {}
for key,df in dict_df.items():
    A = (df['Total Current Assets'] - df['Total Current Liabilities'] )/df['Total Assets']
    B = df['Retained Earnings'] /df['Total Assets']
    C = df['Ebit'] /df['Total Assets']
    D = df['marketCap'] /df['Total Liab']
    E = df['Total Revenue'] /df['Total Assets']
    ratios  = {
        'EBIT Interest Coverage' : df['Ebit'] / df['Interest Expense'],
        'EBITDA Interest Coverage' :(df['Ebit'] + df['Depreciation'])  / df['Interest Expense'],
        'Debt to Equity' : (df['Short Long Term Debt'] + df['Long Term Debt']) /df['Total Stockholder Equity'] ,
        'Debt to Tangible Equity' : (df['Short Long Term Debt'] + df['Long Term Debt']) /(df['Total Stockholder Equity'] - df['Intangible Assets']) ,
        'Cash Flow to Total Debt' : df['Total Cash From Operating Activities'] /(df['Short Long Term Debt'] + df['Long Term Debt']) ,
        'Liabilities to Assets ' : df['Total Liab'] /df['Total Assets'],
        'Cash Ratio' : (df['Short Term Investments'] + df['Cash'] + df['Common Stock']) /df['Total Current Liabilities'] ,
        'Quick Ratio' : (df['Short Term Investments'] + df['Cash'] + df['Common Stock']+ df['Net Receivables'] )/df['Total Current Liabilities'] ,
        'Current Ratio' : df['Total Current Assets'] / df['Total Current Liabilities'],
        'Zscore' : np.dot(1.2,A) + np.dot(1.4,B) + np.dot(3.3,C) + np.dot(0.6,D) + np.dot(1.0,E)
        }
    
    dict_df_ratios[key] = pd.DataFrame.from_dict(ratios)
    
```

```python
for key,df in dict_df_ratios.items():
    solvency = []
    for score in df['Zscore']:
        if score >= 2.4:
            solvency.append(1)
        else:
            solvency.append(0)
    df['solvency'] = solvency
    df.drop('Zscore',axis =1,inplace = True)
    dict_df[key]['solvency'] = solvency
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import cross_val_score , learning_curve
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, classification_report
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
#nos 4 dataframe avec un algrotihme d'imputation
dict_df_ratios.keys()
```

```python
#Methode d'evaluation des modeles
def evaluation (classifier,X_train,y_train,X_test,y_test) : 
    
    modele = classifier.fit(X_train,y_train)
    prediction_base = modele.predict(X_test)
    #Calcul cross validation 10 fold
    score = cross_val_score(classifier,X_train,y_train,scoring = 'f1',cv= 10).mean()
    print("---CV SCORE--- {}".format(score.mean()))
    
    #Classification Report
    print(classification_report(y_test, prediction_base))
    
    #Confusion matrix
    plot_confusion_matrix(modele,X_test, y_test,cmap = 'Blues',display_labels = ['Non solvable','solvable'])
   
```

```python
def plot_roc(X_test,y_test,modele):
    y_score = modele.predict_proba(X_test)[:,1]
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_score)
    J = true_positive_rate - false_positive_rate
    ix = argmax(J)
    best_thresh = thresholds[ix]
    
    #Plot Roc Curve
    plt.subplots(1, figsize=(10,10))
    plt.title('Receiver Operating Characteristic - ')
    plt.plot(false_positive_rate, true_positive_rate,marker = '.',label='ROC curve (area = %0.2f)' % roc_auc_score(y_test, y_score))
    plt.scatter(false_positive_rate[ix], true_positive_rate[ix], marker='o', color='black', label='Best')
    plt.plot([0, 1], ls="--")
    plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc="lower right")
    plt.show()
    print('Best Threshold=%f' % (best_thresh))
    return best_thresh
```

```python
def modele_final(model,X,threshold = 0):
    return model.decision_function(X) > threshold
```
